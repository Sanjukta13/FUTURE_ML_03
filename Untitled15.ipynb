```python
data = pd.read_csv("Resume.csv")
data.head()
```

```python
data.columns
```

```python
resumes = data['Resume_str']
```

```python
job_description = """
Looking for a Data Scientist with skills in Python,
Machine Learning, Data Analysis, SQL,
Deep Learning and Statistics.
"""
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
data['cleaned_resume'] = data['Resume_str'].apply(clean_text)
```

```python
data[['Resume_str','cleaned_resume']].head()
```

```python
all_text = [job_description] + data['cleaned_resume'].tolist()
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(all_text)
```

```python
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
```

```python
data['score'] = similarity_scores[0]
```

```python
data[['Resume_str','score']].head()
```

```python
ranked_resumes = data.sort_values(by='score', ascending=False)
ranked_resumes.head(10)
```

```python
required_skills = [
'python','machine','learning','sql',
'data','analysis','deep','learning'
]
```

```python
def missing_skills(resume):
    words = resume.split()
    missing = [skill for skill in required_skills if skill not in words]
    return missing
```

```python
data['missing_skills'] = data['cleaned_resume'].apply(missing_skills)
```
